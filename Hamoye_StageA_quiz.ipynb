```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
FBS = pd.read_csv("/content/FoodBalanceSheets_E_Africa_NOFLAG.csv",encoding="latin1")

FBS.describe(include="all")
```

```python
#To determine the total sum of Animal Fat Produced in the year 2014 & 2017 using groupby method
FBS.groupby("Item").sum()

#comment: From the above command that was executed, it shows that the total sum of Animal fat produced for the year 2014 & 2017
# are 209460.54 and 269617.53 respectively.
```

```python
#What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?

FBS.describe(include ="all")

#Comment: The mean and standard deviation of the dataset for the year 2015 are 135.236 and 1603.404 respectively
```

```python
#What is the total number and percentage of missing data in 2016 to 2 decimal places?

FBS.isnull().sum()
#Comment: The total number and percentage of a missing data in 2016 are 1535 and 2.52 respectively.
```

```python
#Which year had the highest correlation with ‘Element Code’?
FBS.corr()
#Comment: from the code executed, it shows that Year 2014 has the highest correlatiobn with Element Code
```
